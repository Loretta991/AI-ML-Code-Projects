```python
# from a list
rdd = sc.parallelize([1,2,3])
rdd.collect()
```

```python
rdd = sc.parallelize(('cat', 'dog', 'fish'))
rdd.collect()
```

```python
# from a list of tuple
list_t = [('cat', 'dog', 'fish'), ('orange', 'apple')]
rdd = sc.parallelize(list_t)
rdd.collect()
```

```python
# from a dict
d = {
    'a': 100,
    'b': 200,
    'c': 300
}
rdd = sc.parallelize(d)
rdd.collect()
```
