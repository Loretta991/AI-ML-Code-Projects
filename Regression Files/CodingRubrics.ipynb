```python
import re

def decode(message_file):
    with open(message_file, 'r', encoding='utf-8-sig') as file:
        lines = file.readlines()

    pairs = {}
    for line in lines:
        match = re.match(r'(\d+)\s+(\S+)', line)  # Using regex to match digits and non-whitespace characters
        if match:
            number, word = match.groups()
            pairs[int(number)] = word.strip()

    max_num = max(pairs.keys())
    decoded_message = [pairs[num] for num in range(1, max_num + 1)]
    return ' '.join(decoded_message)

# Using the function with the provided file path
file_path = '/content/sample_data/message_file.txt'
decoded_message = decode(file_path)
print(decoded_message)
```

```python
def decode(message_file):
    with open(message_file, 'r') as file:
        lines = file.readlines()
        data = ' '.join(lines).split()  # Merge lines and split elements by spaces

    pairs = {}
    for i in range(0, len(data), 2):
        number = int(data[i])
        word = data[i + 1]
        if number in [1, 3, 6]:
            pairs[number] = word

    decoded_message = ' '.join([pairs[i] for i in [1, 3, 6]])
    return decoded_message

file_path = '/content/sample_data/message_file.txt'
decoded_message = decode(file_path)
print(decoded_message)
```
