```python


!pip install pyttsx3 SpeechRecognition

import pyttsx3
import speech_recognition as sr

def speak(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

def listen():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening...")
        audio = recognizer.listen(source)
    try:
        command = recognizer.recognize_google(audio)
        print(f"User said: {command}")
        return command
    except sr.UnknownValueError:
        speak("Sorry, I did not understand that.")
        return None
    except sr.RequestError:
        speak("Sorry, my speech service is down.")
        return None

def get_numeric_input(prompt):
    while True:
        speak(prompt)
        user_input = listen()
        if user_input:
            try:
                return float(user_input)
            except ValueError:
                speak("Please provide a valid number.")

def cd_analysis_chatbot():
    speak("Welcome to the CD Maturity Analysis Chatbot!")

    principal = get_numeric_input("Please say the initial deposit amount in dollars.")
    annual_rate = get_numeric_input("Please say the annual interest rate in percent.")
    years = get_numeric_input("Please say the duration of the CD in years.")

    maturity_value = principal * (1 + (annual_rate / 100)) ** years
    speak(f"The maturity value of your CD after {years} years will be {maturity_value:.2f} dollars.")

    speak("Would you like to reinvest the matured amount or withdraw it? Please say reinvest or withdraw.")
    choice = listen()
    if choice:
        if "reinvest" in choice.lower():
            speak("Reinvesting your matured amount could lead to further growth!")
        elif "withdraw" in choice.lower():
            speak("You can withdraw the matured amount for personal use or other investments.")
        else:
            speak("Invalid choice. Please decide wisely upon maturity.")

    speak("Thank you for using the CD Maturity Analysis Chatbot!")

# Run the chatbot
cd_analysis_chatbot()
```
