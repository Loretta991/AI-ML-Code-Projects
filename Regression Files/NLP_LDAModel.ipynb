```python
from gensim import corpora
from gensim.models import LdaModel
from gensim.models import CoherenceModel
from pprint import pprint

# Sample documents
documents = [
    "The economy is experiencing a downturn",
    "Climate change is a global concern",
    "Political instability is rising in the region"
]

# Create dictionary and corpus
dictionary = corpora.Dictionary(documents)
corpus = [dictionary.doc2bow(doc) for doc in documents]

# Set hyperparameters
num_topics = 3  # Number of topics
alpha = 'auto'  # Automatic setting of alpha
beta = 'auto'   # Automatic setting of beta

# Build LDA model
lda_model = LdaModel(corpus=corpus, id2word=dictionary, num_topics=num_topics, alpha=alpha, eta=beta)

# Print topics and their top words
pprint(lda_model.print_topics())

# Calculate coherence score (optional)
coherence_model = CoherenceModel(model=lda_model, texts=documents, dictionary=dictionary, coherence='c_v')
coherence_score = coherence_model.get_coherence()
print("Coherence Score:", coherence_score)
```
