```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = "/content/sample_data/Mag6PlusEarthquakes_1900-2013.xlsx"
data_frame = pd.read_excel(file_path)

# Define a function to split place into place and region
def split_place(place):
    if pd.isna(place):
        return "", ""

    parts = place.split(',')
    if len(parts) >= 2:
        return parts[0].strip(), parts[1].strip()
    return place, ""

# Apply the function to split place into new columns
data_frame[['place', 'region']] = data_frame['place'].apply(split_place).apply(pd.Series)

# Group by region and place, then count the earthquakes in each group
grouped_data = data_frame.groupby(['region', 'place']).size().reset_index(name='count')

# Define a dictionary mapping continents to their respective regions
continent_regions = {
    'North America': ['Alaska', 'California', 'Japan', 'Mexico'],  # Add more regions as needed
    'Europe': ['Italy', 'Greece'],  # Add more regions as needed
    # Add more continents and regions as needed
}

# Define a dictionary to map years to their respective ranges
year_ranges = {
    range(1900, 1931): '1900-1930',
    range(1940, 1971): '1940-1970',
    range(1980, 2001): '1980-2000',
    range(2001, 2015): '2001-2014',
}

# Iterate through each continent and create separate plots
for continent, regions in continent_regions.items():
    for region in regions:
        region_data = data_frame[data_frame['region'] == region]

        # Group data by year ranges
        region_data['year'] = region_data['Date'].dt.year
        region_data['year_range'] = region_data['year'].apply(lambda x: next((v for k, v in year_ranges.items() if x in k), 'Unknown'))
        year_grouped = region_data.groupby('year_range').size().reset_index(name='count')

        # Create a bar plot for earthquake counts by year range
        plt.figure(figsize=(8, 4))
        plt.bar(year_grouped['year_range'], year_grouped['count'], color='dodgerblue')
        plt.xlabel('Year Range')
        plt.ylabel('Number of Earthquakes')
        plt.title(f'{continent} - Earthquake Counts by Year Range')
        plt.tight_layout()
        plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import numpy as np

# Load the dataset
file_path = "/content/sample_data/Mag6PlusEarthquakes_1900-2013.xlsx"
data_frame = pd.read_excel(file_path)

# Define a function to split place into place and region
def split_place(place):
    if pd.isna(place):
        return "", ""

    parts = place.split(',')
    if len(parts) >= 2:
        return parts[0].strip(), parts[1].strip()
    return place, ""

# Apply the function to split place into new columns
data_frame[['place', 'region']] = data_frame['place'].apply(split_place).apply(pd.Series)

# Define a dictionary mapping continents to their respective regions
continent_regions = {
    'North America': ['Alaska', 'California', 'Japan', 'Mexico'],  # Add more regions as needed
    'Europe': ['Italy', 'Greece'],  # Add more regions as needed
    # Add more continents and regions as needed
}

# Create a heatmap for earthquake magnitudes by region
for continent, regions in continent_regions.items():
    region_data = data_frame[data_frame['region'].isin(regions)]

    heatmap_data = region_data.pivot_table(index='region', columns='year', values='mag', aggfunc='mean')

    plt.figure(figsize=(10, 6))
    sns.heatmap(heatmap_data, cmap='YlOrRd', annot=True, fmt=".2f", linewidths=0.5)
    plt.title(f'{continent} - Earthquake Magnitude Heatmap by Region')
    plt.tight_layout()
    plt.show()

# Create a scatter cluster plot for magnitude vs. depth by region
for continent, regions in continent_regions.items():
    region_data = data_frame[data_frame['region'].isin(regions)]

    cluster_data = region_data[['mag', 'depth']]
    kmeans = KMeans(n_clusters=4, random_state=0).fit(cluster_data)
    region_data['cluster'] = kmeans.labels_

    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=region_data, x='mag', y='depth', hue='cluster', palette='tab10')
    plt.xlabel('Magnitude')
    plt.ylabel('Depth')
    plt.title(f'{continent} - Magnitude vs. Depth by Region (Clustered)')
    plt.tight_layout()
    plt.show()

# Create a pie chart for earthquake counts by region
for continent, regions in continent_regions.items():
    region_counts = data_frame[data_frame['region'].isin(regions)]['region'].value_counts()

    plt.figure(figsize=(8, 6))
    plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
    plt.title(f'{continent} - Earthquake Counts by Region (Pie Chart)')
    plt.tight_layout()
    plt.show()
```
