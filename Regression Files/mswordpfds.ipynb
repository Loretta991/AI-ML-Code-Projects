```python
!pip import os
!pip import docx
!pip from docx2pdf 
!pip import convert

import os
import subprocess
!pip from docxcompose.composer import Composer
!pip from docxtpl import DocxTemplate
!pip from docx2pdf import convert

# Install required libraries
subprocess.call(["pip", "install", "pandas"])
subprocess.call(["pip", "install", "docxcompose"])
subprocess.call(["pip", "install", "docxtpl"])

# Ask user for file name on desktop
desktop_path = os.path.join(os.path.expanduser("~"), "Desktop")
filename = input("Enter file name on desktop: ")
file_path = os.path.join(desktop_path, filename)

# Check if file exists
if not os.path.isfile(file_path):
    print("File does not exist.")
else:
    # Read text from file
    with open(file_path, "r") as f:
        text = f.read()

    # Create a new Word document from a template
    template = DocxTemplate("template.docx")
    context = {"text": text}
    template.render(context)
    template.save("output.docx")

    # Convert the Word document to PDF
    convert("output.docx")

    # Ask user what type of save as file they want
    file_type = input("Enter file type (PDF or DOCX): ")

    # Save the file as PDF or DOCX
    if file_type.upper() == "PDF":
        os.rename("output.pdf", os.path.join(desktop_path, "output.pdf"))
    elif file_type.upper() == "DOCX":
        os.rename("output.docx", os.path.join(desktop_path, "output.docx"))
    else:
        print("Invalid file type.")
```
