```python
import pandas as pd

# Load the Excel file into a pandas DataFrame
excel_file_path = '/content/sample_data/fndds_ingredient.xlsx'
df = pd.read_excel(excel_file_path)

# Split the "code" column into two separate columns
df[['code_numeric', 'Ingredient description']] = df['code'].str.extract('(\d+),"(.*)"')

# Remove quotes from all columns
df = df.apply(lambda x: x.str.strip('"'))

# Drop the original "code" column
df.drop(columns=['code'], inplace=True)

# Rename the columns to match your desired field names
df.columns = ["code_numeric", "Ingredient description", "Nutrient code", "Nutrient value", "Nutrient value source", "FDC ID", "Derivation code", "SR AddMod year", "Foundation year acquired", "Start date", "End date"]

# Save the modified DataFrame as a new Excel file
output_excel_file = 'fndds_ingredients_split.xlsx'
df.to_excel(output_excel_file, index=False)

# Display the first 10 records
print(df.head(10))

print(f'Modified data has been saved to "{output_excel_file}"')
```
