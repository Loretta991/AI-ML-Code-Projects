```python
import pandas as pd
import numpy as np
from faker import Faker
import random
import datetime

fake = Faker()

# Generate dummy data for California Veterans Administration (CALVET)
np.random.seed(42)

# Create a DataFrame with dummy data
data = {
    'VeteranID': range(1, 101),
    'FirstName': [fake.first_name() for _ in range(100)],
    'LastName': [fake.last_name() for _ in range(100)],
    'Gender': np.random.choice(['Male', 'Female'], size=100),
    'AgeAtEnlistment': np.random.randint(18, 40, size=100),
    'EnlistmentDate': [fake.date_between(start_date='-50y', end_date='-10y') for _ in range(100)],
    'ServiceBranch': np.random.choice(['Army', 'Navy', 'Air Force', 'Marines', 'Coast Guard'], size=100),
    'EnlistmentLocation': [fake.city() for _ in range(100)],
    'ServiceDurationYears': np.random.randint(1, 30, size=100),
    'DischargeDate': [fake.date_between(start_date='-5y', end_date='today') for _ in range(100)],
    'DischargeStatus': np.random.choice(['Honorable', 'General', 'Dishonorable'], size=100),
    'ReceivingBenefits': np.random.choice([True, False], size=100),
    'BenefitType': np.random.choice(['Education', 'Healthcare', 'Housing', 'Pension'], size=100),
    'InjuredDuringDuty': np.random.choice([True, False], size=100),
    'InjuryLocation': np.random.choice(['WW2', 'Vietnam', 'Korea', 'Afghanistan', 'Iraq', 'Other'], size=100),
    'InjuryDate': [fake.date_between(start_date='-30y', end_date='-1y') if injured else None for injured in np.random.choice([True, False], size=100)],
    'BurialFacility': [fake.city() for _ in range(100)],
    'Rank': [fake.random_element(['Private', 'Corporal', 'Sergeant', 'Lieutenant', 'Captain', 'Major', 'Colonel', 'General']) for _ in range(100)]
}

calvet_data = pd.DataFrame(data)

# Save the data to a CSV file
calvet_data.to_csv("calvet_data.csv", index=False)
```
