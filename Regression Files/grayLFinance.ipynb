```python
import yfinance as yf
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd

def download_historical_prices(symbols, start_date, end_date):
    prices = pd.DataFrame()
    for symbol in symbols:
        stock = yf.Ticker(symbol)
        daily_prices = stock.history(start=start_date, end=end_date)
        daily_prices.columns = [f'{symbol}_{col}' for col in daily_prices.columns]
        prices = pd.concat([prices, daily_prices], axis=1)
    return prices
symbols = ['AAPL', 'MSFT']
start_date = '2022-01-01'
end_date = '2022-12-31'

prices = download_historical_prices(symbols, start_date, end_date)
print(prices)
   

# Download historical prices for AAPL from Yahoo Finance
symbol = 'AAPL'
stock = yf.Ticker(symbol)
data = stock.history(period='max')

# Compute daily returns as percentage change
returns = data['Close'].pct_change().dropna()

# Create a histogram with 20 bins
plt.hist(returns, bins=20)

# Add a title and axis labels
plt.title(f'Histogram of Daily Returns for {symbol}')
plt.xlabel('Daily Return')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```
