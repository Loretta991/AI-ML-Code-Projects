```python
import requests
from bs4 import BeautifulSoup

url = "https://www.indeed.com/resumes"
params = {
    "q": "Python",
    "l": "Silicon Valley, CA"
}

response = requests.get(url, params=params)
soup = BeautifulSoup(response.content, "html.parser")

resume_listings = soup.find_all("div", class_="rezemp-u")
count = 0

for listing in resume_listings:
    count += 1
    print(f"Resume {count}:")
    print("------------------------")
    print(listing.get_text().strip())
    print()

    if count == 20:
        break
```

```python
import requests
from bs4 import BeautifulSoup

url = "https://www.indeed.com/resumes"
params = {
    "q": "Python",
    "l": "San Jose CA"
}

response = requests.get(url, params=params)
soup = BeautifulSoup(response.content, "html.parser")

resume_listings = soup.find_all("a", class_="app_link")
count = 0

for listing in resume_listings:
    count += 1
    if count > 5:
        break

    print(f"Resume {count}:")
    print("------------------------")
    print(listing.get_text().strip())
    print()
```
