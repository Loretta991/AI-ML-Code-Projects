```python

#The first line imports the numpy and matplotlib.pyplot libraries.
import numpy as np
import matplotlib.pyplot as plt
import urllib.request
```

```python

#The next few lines define two functions: read_mnist_data and read_mnist_labels. 
#These functions are used to read in the MNIST 
#data and labels from a URL.
# Define a function to read in the MNIST data from a URL

def read_mnist_data(url):
    with urllib.request.urlopen(url) as response:
        data = response.read()
    return np.frombuffer(data, np.uint8, offset=16).reshape(-1, 784)
```

```python
#The read_mnist_data function takes a URL as input, downloads the data from the 

#URL using urllib.request.urlopen, and then converts the data to a numpy array of
#unsigned 8-bit integers (np.uint8) with an offset of 16 bytes (the first 16 bytes 
#are header information). The function then reshapes the data into a 2D array 
#with 784 columns (one for each pixel) and an unknown number of rows 
#(corresponding to the number of images in the dataset).

# Define a function to read in the MNIST labels from a URL
def read_mnist_labels(url):
    with urllib.request.urlopen(url) as response:
        data = response.read()
    return np.frombuffer(data, np.uint8, offset=8)
```

```python


#The read_mnist_labels function is similar to read_mnist_data, but it reads in the labels instead of the 
#image data. The label data starts at an offset of 8 bytes and consists of unsigned 8-bit integers.
#The next two lines define the URLs for the MNIST data and labels.


# URLs for the MNIST data and labels
url_data = "http://yann.lecun.com/exdb/mnist/train-images-idx3-ubyte.gz"
url_labels = "http://yann.lecun.com/exdb/mnist/train-labels-idx1-ubyte.gz"
```

```python
#The next two lines read in the MNIST data and labels using the read_mnist_data and read_mnist_labels functions, respectively.
# Read in the MNIST data and labels


def read_mnist_labels(url):
    # Open the URL and read the data
    with urllib.request.urlopen(url) as response:
        data = response.read()

    # Check that the size of the downloaded data matches the expected size for MNIST labels
    expected_size = 60000 + 8
    if len(data) != expected_size:
        raise ValueError(f"Unexpected data size ({len(data)}) for MNIST labels")

    # Return the data as a NumPy array with shape (60000,)
    return np.frombuffer(data, np.uint8, offset=8)   

X_train = read_mnist_data(url_data)
y_train = read_mnist_labels(url_labels)
```

```python

#The final two lines print the shape of the data and labels to confirm that they have been loaded correctly.

# Print the shape of the data and labels
print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```
