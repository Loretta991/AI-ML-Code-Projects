```python
#9. Program to determine if two list are circulary identical
# example: list1 = [1,2,3] list2 = [3,2,1]
cir1 = [1,2,3]
cir2 = [3,2,1]  

print('input strings : ',  str(cir1) , str(cir2))
# sort the two strings
cir3 = cir1.sort()
cir4 = cir2.sort()
if cir3 == cir4:  
  print('output: identical')
else:
  print('not identical')
```
